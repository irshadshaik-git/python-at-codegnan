```python
#check which one is even and which one is odd if the num is even squre it if the number iss odd then add +1
lst=list(map(int,input().split()))
for num in lst:
    if num%2==0:
        print(num**2)
    else:
        print(num+1)
```

```python
lst=list(map(int,input().split()))
for num in lst:
    if num%3==0 and num%5==0:
        print("FIZZBIZZ")
    elif num%3==0:
        print("FIZZ")
    elif num%5==0:
        print("BIZZ")
    else:
        print(num)
        
```

```python
lst=list(map(int,input().split()))
res=[]
for num in lst:
    if num%3==0 and num%5==0:
        res.append("FIZZBIZZ")
    elif num%3==0:
        res.append("FIZZ")
    elif num%5==0:
        res.append("BIZZ")
    else:
        res.append(num)
print(res)
```
