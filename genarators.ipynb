```python
# return 1 to n numbers using functions
import sys
def num(n):
    res=[]
    for i in range(1,n+1):
        res.append(i)
    print(res)
r=num(10)
print(sys.getsizeof(r))
print(r)
         
```

```python
import sys
def gen(n):
    res=[]
    for i in range(1,n+1):
        yield i
r=gen(3)
print(r)
print(sys.getsizeof(r))
print(r)
```

```python
# return 1 to n numbers using functions
import sys
def num(n):
    res=[]
    for i in range(1,n+1):
        res.append(i**3)
    print(res)
r=num(10)
print(sys.getsizeof(r))
print(r)
         
```
