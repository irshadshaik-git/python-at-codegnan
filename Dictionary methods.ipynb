### Dictionary Methods

- get(key,default_value):It returns the value if key is dictionary,otherwise it returns defaul value

```python
# dictionary key ate uniue
details={"Age":22,"name": "babu","course":"python","Batch":57,"Age":20}
print(details['name'])
#print(details['marks']) # This abe raises KeyError
print(details.get('name'))
print(details.get('marks'))#get Method bydefault it returns None if key not exists in details
print(details.get('marks',"marks are not available"))#it returns defualt value ,because
```

- update(new_dictionary) : This methods updates the values if new_dictionary keys already exits in Dictionary,otherwise it adds new items to dictionary

```python
details={"Age":22,"name": "babu","course":"python","Batch":57}
details.update({'batch':58,'course':'java',"Location":"hyd"})
print(details)
```

- Keys(): This methods returns the list dictionary Keys
- value():This methods returns the list dictionary values
- items():This methods returns the list  of Tuple of key_value pairs  

```python
details={'Age': 22, 'name': 'babu', 'course': 'java', 'Batch': 57, 'batch': 58, 'Location': 'hyd'}
keys=details.keys()
print(type(keys))
print(keys)
print(details.values())
print(details.items())
```

- popitem(): it return and removes the list item as tuple
- pop(key): this method removes the specified key item and return the specified key's
  value,otherwise 'KeyError'

```python
details={'Age': 22, 'name': 'babu', 'course': 'java', 'Batch': 57, 'batch': 58, 'Location': 'hyd'}
item=details.popitem()
print(item)
print(details)
val= details.pop('name')
print(val)
print(details)
```
