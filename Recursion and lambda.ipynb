```python
#factorial by recursive
def factorial(x):
    if x==0 or x==1:
        return 1
    return x*factorial(x-1)
print(factorial(5))
```

```python
n = int(input("Enter the number: "))
i=1
sum=0
while(i<=n):
    sum+=i
    i+=1
print(sum)
```

```python
#sum of the digits
def sum(x):
    if x==0 or x==1:
        return 0
    return x % 10 + sum(x//10)
print(sum(465))
```

```python
# find the squeres of the list of integers using map and lambda
arr=[1,2,3,4,5]
print(list(map(lambda x: x**2 , arr)))
```

```python
# check weather the number is even or odd using lambda
n=int(input())
even_or_odd=lambda x: "even" if x%2==0 else "odd"
print(even_or_odd(n))

```

```python
#find the frequency of each charecter
string=input()
count={}
for char in string:
    if char in count:
        count[char]+=1
    else:
        count[char]=1
print(count)
```
