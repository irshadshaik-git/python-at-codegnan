`Class` :
- class is a blue print of a realworld object or template of the object
- class is a collection altributes (ie.data)and methode(ie.behavior)

`syntax`: class class_name:# attributes #methodes


`object`:
- object is an instance of a class and realworld item

`syntax`:
#obect reacton
    #obj=class_name(arguments)

```python
# difine persion class
class student:
    school="xyz school" # class data attributes
    def __init__(self, name, age): # Initialization Constructor (Magic method)
        self.name = name # instance attribute
        self.age = age # instance attribute
        print("this is student calss__init__method")
    def details(self):
        return f"My name is {self.name} and i am a {self.age} years old"
s1 = student("ravi",20)
s2 = student("jannu",18)
print(s1.name)
print(s2.name)
print(s1.age)
print(s2.age)
print(s1.details(),f"he is in the {s1.school}")
print(s2.details())
    
```
