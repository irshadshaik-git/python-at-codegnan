```python
class Suspect:
    def __init__(self, name, statement, guilty=False):
        self.name = name
        self.statement = statement
        self.__guilty = guilty      # Private attribute

    def talk(self):
        print(f"\n{self.name} says:")
        print(self.statement)

    def is_guilty(self):
        return self.__guilty


class Detective:
    def __init__(self, name):
        self.name = name
        self.clues = []

    def collect_clue(self, clue):
        self.clues.append(clue)
        print("\nClue Collected Successfully!")

    def show_clues(self):
        print("\n===== COLLECTED CLUES =====")
        if len(self.clues) == 0:
            print("No clues collected.")
        else:
            for i, clue in enumerate(self.clues, 1):
                print(f"{i}. {clue}")


# Create Detective
detective = Detective(input("Enter Detective Name: "))

# Create Suspects
suspect1 = Suspect(
    "Ravi",
    "I was at the office until 9 PM.",
    False
)

suspect2 = Suspect(
    "Priya",
    "I was shopping with my friends.",
    False
)

suspect3 = Suspect(
    "Karan",
    "I don't know anything about the robbery.",
    True
)

while True:

    print("\n========== DETECTIVE MENU ==========")
    print("1. Visit Crime Scene")
    print("2. Talk to Ravi")
    print("3. Talk to Priya")
    print("4. Talk to Karan")
    print("5. Show Collected Clues")
    print("6. Accuse Suspect")
    print("7. Exit")

    choice = input("\nEnter Choice: ")

    if choice == "1":
        detective.collect_clue(
            "A black glove was found at the crime scene."
        )

    elif choice == "2":
        suspect1.talk()
        detective.collect_clue("Ravi has an office attendance record.")

    elif choice == "3":
        suspect2.talk()
        detective.collect_clue("Priya has shopping receipts.")

    elif choice == "4":
        suspect3.talk()
        detective.collect_clue("Karan's fingerprints match the glove.")

    elif choice == "5":
        detective.show_clues()

    elif choice == "6":
        print("\nChoose Suspect")
        print("1. Ravi")
        print("2. Priya")
        print("3. Karan")

        accuse = input("Enter Choice: ")

        if accuse == "1":
            if suspect1.is_guilty():
                print("\nCongratulations! Ravi is guilty.")
            else:
                print("\nWrong Accusation! Ravi is innocent.")

        elif accuse == "2":
            if suspect2.is_guilty():
                print("\nCongratulations! Priya is guilty.")
            else:
                print("\nWrong Accusation! Priya is innocent.")

        elif accuse == "3":
            if suspect3.is_guilty():
                print("\n🎉 Congratulations Detective!")
                print("You solved the mystery.")
                print("Karan is the criminal.")
            else:
                print("\nWrong Accusation!")

    elif choice == "7":
        print("\nThank you for playing!")
        break

    else:
        print("\nInvalid Choice!")
```
