 ### set characteristics
 

```python
# set unordered collection
#set returns unique elements
s={1,2,3,2,3,'hii','a',2.02,(10,20,30)}
print(s)
#print(s[0])# set does not support indexing
```

```python
#List can't be used as set element
s={1,2,[3,2,3],'hii','a',2.02,(10,20,30)}
print(s)
```

### set methods

```python
# Element adding mathod
# Add(element): it adds the element to set if element not in set, otherwise nothing
s={1,2,3,4,5,4,3}
print(s)
s.add(100)
s.add(5)
print(s)
```

```python
# Element removing methods
# remove(item): it remove the item if it exits in set otherwise,'KeyError'
# discard(item):it remove the item if it exits in set otherwise nothing
```

```python
s={1,2,3,4,5,4,3}
s.remove(5)
print(s)
# s.remove(100)# KeyError
s.discard(3)
print(5)
s.discard(100)
print(s) # it 
```

```python
### sets Operations
# union(set): it returns the all unique element from all sets
s1={1,3,5,7}
s2={4,5,6,7}
s3={1,2,8,9}
union=s1.union(s2, s3)
print(union)
res = s1|s2|s3
print(res)
```

```python
### sets Operations
# intersection(set): it returns the common value element from all sets
s1={1,3,5,7}
s2={4,5,6,7}
s3={5,1,2,8,9}
intersection=s1.intersection(s2, s3)
print(intersection)
res = s1|s2|s3
print(res)
```

```python
### sets Operations
# difference(set): it returns the a set elements those are not in another sets
s1={1,3,5,7}
s2={4,5,6,7}
s3={5,1,2,8,9}
intersection=s1.intersection(s2, s3)
print(intersection)
res = s1-s2-s3
print(res)
print((s1 | s2)-s3)
```

```python
### sets Operations
# symmertric_difference(set): it returns the non commrn element from all sets
s1={1,3,5,7}
s2={4,5,6,7}
s3={5,1,2,8,9}
st_diff= s1.symmetric_difference(s2)
print(st_diff)
print(s1 ^ s2 ^ s3)
```

```python
# issubset(set)
# issuperset(set)
# isdisjoint(set)
s1={1,3,5,7}
s2={1,3,5,7}

a=s1.issubset(s2)
print(a)
b=s2.issuperset(s1)
print(b)
c=s1.isdisjoint(s2)# if the set has not common elements
print(c)
```
