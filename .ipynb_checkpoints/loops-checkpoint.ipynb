```python
a,b,c=map(int,input().split())
if a%2==0:
    print("even")
else:
    print("odd")
if b%2==0:
    print("even")
else:
    print("odd")
if c%2==0:
    print("even")
else:
    print("odd")
```

```python
#check which one is even and which one is odd if the num is even squre it if the number iss odd then add +1
lst=list(map(int,input().split()))
for num in lst:
    if num%2==0:
        print(num**2)
    else:
        print(num+1)
```

```python
lst=list(map(int,input().split()))
for num in lst:
    if num%3==0 and num%5==0:
        print("FIZZBIZZ")
    elif num%3==0:
        print("FIZZ")
    elif num%5==0:
        print("BIZZ")
    else:
        print(num)
        
```

```python
lst=list(map(int,input().split()))
res=[]
for num in lst:
    if num%3==0 and num%5==0:
        res.append("FIZZBIZZ")
    elif num%3==0:
        res.append("FIZZ")
    elif num%5==0:
        res.append("BIZZ")
    else:
        res.append(num)
print(res)
```

```python
lst=list(map(int,input().split()))
target=int(input())
n=len(lst)
for i in range(n):
    if lst[i]==target:
        print(i)
        break;
else:
    print(-1)
```

```python
# prime numbers
n=int(input())
factor=list()
for i in range(1,n+1):
    if n%i==0:
        factor.append(i)
print(factor)
    
```

```python
# prime number2
n= int(input())
for i in range (2,n//2+1):
    if n%i==0:
        print(f"{n}not prime")
        break;

else:
    print(f"{n} is prime")
```
