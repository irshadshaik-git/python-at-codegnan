### remove white space or pattern at ends of string
 - strip(): by default it removes the white space at both ends of string
 - lstrip(): it removes the white space at left side of string
 - rstrip(): it remove the white space at right side of string


```python
s= ""
print(s)
print(s.lstrip())
print(s.rstrip())
```

```python
s= "#####eeshu######"
print(s)
print(s.lstrip("#"))
print(s.rstrip("#"))
```

```python
#slipt(): it splits the entrire string into list of substring based on the specifide seperator by default it splits at space
string = "the sky is blue and ocean is blue"
print(string.split())
print(string.split("is"))
print(string.split('e'))
```

### substring index finding
- index(substring):it substring is exists in string, it returns the first occurencce of substring position,otherwise ValueError
- find(substring):if substrings is exits in string,it returns the first occrencce of substring position,otherwise -1



```python
string = "the sky is blue and ocean is blue"
print(string.find('is'))
print(string.find('z'))
print(string.find('is',10))# it start checking from 10th position
```

```python
#count(substring): it returns the total non-overlapping occucrence of substring
s="ABCBCB"
print(s.count("B"))
print(s.count("BCB"))
```

```python
s="ABCBCBbababb"
print(s.count("B"))
print(s.count("BCB"))
lower_string=s.lower()
print(lower_string)
print(lower_string.count('b'))
print(s.upper().count('B'))
print(s.count('z'))
```

```python
#replace(old_stbstring,new_substring): it replace the old sub-string with new sub string
string = "the sky is blue and ocean is Blue"
print(string.replace("blue","RED"))
print(string.replace('e','E'))
print(string.replace('e','E',2))#it replace the first 2 occurance
```

### string checking methods
- string checking methods returns the boolean value as result
- isupper(): it returns True, If entire string is in upper case
- islower():it returns True, If entire string is in lower case
- isalpha():it returns True, If entire string is in contains alphabets
- isdegit(): it return true, if entire string contains digits
- isalnum(): it returns True, If entire string contains either alphabets or digits
- startswith(substring):it returns True, If entire string-start with substring
- endswith(substring):it returns True, If string ens with substring

```python
print("CODE".isupper())
print("CODE".islower())
print("code".islower())
print('1234'.isdigit())
print('123abfdsg'.isdigit())
print('code'.isalpha())
print('code123'.isalnum())
print('1234'.isalnum())
print("code".startswith("c"))
```
