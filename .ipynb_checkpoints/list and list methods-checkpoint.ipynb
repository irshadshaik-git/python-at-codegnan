### list characteristics

```python
#Difine list
#list is ordered collection
lst=[1,2,3,4,5,"hi","hello",1,3,4]
print(lst)
```

```python
# list allows indexing and slicing
# Accessing List items
lst=[1,2,3,4,5,"hii","Heloo",1,3,4]
print(lst[6])
print(lst[-2])
```

```python
lst=[10,20,30,40,50,60,70]
print(lst[:5:-1])
print(lst[:500])
print(lst[::-1])#revers list
```

```python
# empty list creation
l1=[]
l2=list()
print(type(l1),type(l2))
```

```python
#list operation
#1. list concatination(using + operation)
#2.list repetation (using * operator)
l1=[1,2,3,]
l2=[3,4,5,6,7]
print(l1+l2)
print(l1*4)
print([0]*10)
```

### list methods
 - item adding methods
     - append(item):it adds item at end of list
     - extend(iterable) : it adds the all iterator elements at end of list
     - insert(pos,item) : it adds item at specified position


```python
l1=[10,20,30,40]
l1.append(100)# you can add a single element
print(l1)
l1.extend([1,2,3,4])
print(l1)
l1.insert(0,1000)
print(l1)
```

- item removing methods
  - remove(item): it remove the  first occurence item from the list if it is exists in list othervise ValueError
  - pop(pos): it remove the specific position items if position is passed
  - clear(): it clear all the elements

```python
lst=[10,20,30,40,50,20,]
val=lst.pop()
print(val)
print(lst)
val2=lst.pop(0)# it removes the 0th position
print(val2)
print(lst)
lst.clear()
print(lst)
```

```python
# index and count()
# index(item): if item is present in list it returns first occurance the index otherwise
# count(item): it returns the count of Total occurance of items
lst=[10,20,30,40,50,20,20]
ind=lst.index(20)
print(ind)
print(lst.count(20))
```

```python
# sorting items
#sort(reverse=flase): by default it sorts the origina list into ascending order
#        if reverser is true it sorts inDecending order
lst =[1,4,3,5,-5,3,-45,-4,30]
lst.sort()
print(lst)
lst.sort(reverse=True)
print(lst)
```

```python
# Buitin Functions, len(), min(),max(), sum()
lst=[1,2,3,4,5]
print(len(lst),min(lst),max(lst),sum(lst)) 
print("average",sum(lst)/len(lst))
```
