### Methods

- Methods: the function defination is implemented inside the class those function called as Methods.
  - Types of Methods
    1. instance Method
    2. class Method
    3. static Method
- 1. instance method
  - the method defination takes first parameter as "self" thoes method as `Instance method` . this methods works on instance attributes.
      `syntax` <br>
          `          class class_name:`                   <br>
          `              def method_name(self):`           <br>
          `                    #block code`
- class method: The method defination starts with`classmethod` and decorator and it takes first parameter as `cls` those methods called as `class methods`. This methods works on class Attributes.
  `syntax:`<br>
         `           class class_name:`              <br>
          `                    @classmethod`          <br>
          `                def method_name(cls):`       <br>
          `             #block code`
  - static method: the method defination starts with `staticmathod` decorator and it dosen't take first parameter as `self or cls` those methods called as `static methos`. This mwthods work on except instance attribets and attributes.`syntax:`
 
     `           class class_name:`              <br>
          `                    @staticmethod`          <br>
          `                def method_name(parameters):`       <br>
          `             #block code`

   - inistialization(__init__)constracter ; This is used for to inistilized the instance attributes data and its constaction takes first parameter as `self` and this constracter exicutes whenever an object is creats for the class
 

```python
#class attributs - school
# instance attributes - id, name, class
class student:
    
    school="xyz student"
    def __init__(self,stdid, name,standerd):
        self.name = name
        self.stdid = stdid
        self.standerd = standerd
    def student_details(self):
        return f"I MY self {self.name} and from {self.standerd} class"
s1=student(101,"ravi",9)
print(s1.name)
print(f"i study in {s1.school}")
print(s1.student_details())
```

```python
# instance mathod : 
#class attributs - school
# instance attributes - id, name, class
class student:
    
    school="xys codegan"  # class atributes
    def __init__(self,stdid, name,standerd):
        self.name = name # instance attributes
        self.stdid = stdid  # instance attributes
        self.standerd = standerd   # instance attributes
    def student_details(self):    # instance method
        return f"I MY self {self.name} and from {self.standerd} class"
    # class method
    @classmethod
    def update_school(cls, new_school_name):
        cls.school = new_school_name
        return "school name Updated"
    @staticmethod
    def percentage(marks:list):
        return (sum(marks)/300) * 100
        

s1=student(101,"ravi",9)
s2=student(102,"raju",10)
print(s1.school,s2.school)
print(s1.student_details())
print(s1.update_school("codegan school"))
print(s1.school, s2.school)
print(s2.student_details())

```

### Encapsulation:
       -> Encapsulation means, wraping the methods and atrribuites in single unit that one is class. the main aim Encapsulation is restrict the direct acces of attribution and methods.
       ->This type of Retriction can implement by using `access Moifiers`
       -> types of acess modifiers
          1. Public
          2. protected
          3. private
          -> in python we don't have any Keywords to represent it public, protected and private,but by using underscore(_) can define
          1. public : The Public Members(i.e attributes + method) doesn't starts with under score. This public members can access any where in the program.
          2. Protected : The Public Members(i.e attributes + method)  starts with single  under score. This members can access in inside the class and in subclasses also.
          3. private ;  The Public Members(i.e attributes + method)  starts with double under score(__). This members can access in only inside.
        

```python
class Bank:
    def __init__(self, name, account, password, balance):
        self.name = name
        self.account = account
        self.password = password
        self.balance = balance
    def __total_user(self): # getter method
        return "total users : 101"
    def get_balance(self): # getter method
        return f"Current Balance is : {self._balance}"
    def deposite(self, deposite_amount): # setter method
        self._balance += deposite_amount
        return f"{deposite_amount} Deposite successfully"
        
u1=Bank("ravi", 1001, "ravi@123", 10000)

print(u1.name)
print(u1.deposite(5000))
print(u1.get_balance())
print(u1._balance) # Protecting members access in outside the class
print(u1._bank__total_users()) # Name Mangling:syntax: obj._classname__attribute
```
