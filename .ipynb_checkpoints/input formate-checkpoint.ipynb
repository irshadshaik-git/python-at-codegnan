### input Formating

```python
#reading integer from user
num = int(input("Enter the number: "))
print(type(num))
print(num)
```

```python
#reading float value from user
val=float(input("enter the nnumber : "))
print(type(val))
print(val)
```

```python
# Complex reading from user
val = complex(int(input()),int(input()))
print(type(val))
print(val)
```

```python
# boolean 
val = bool(input())
print(type(val))
print(val)
```

```python
val = bool(int(input()))
print(type(val))
print(val)
```

```python
# reading string value from user
st=str(input())
print(type(st))
print(st)
```

```python
# reading list of integers from user
lst=input().split()
print(lst)
```

```python
lst=map(int,input().split())
print(lst)
```

```python
lst=list(map(int,input().split()))
print(lst)
```

```python
lst=list(map(str,input().split()))
print(lst)
```

```python
#approuch 2
lst=input().split()
print(lst)
```

```python
# reading Tuple of integers from user
lst=tuple(map(str,input().split()))
print(lst)
```

```python
# eval(): This function evluates the string type expression
val = eval(input())
print(type(val))
print(val)
```
